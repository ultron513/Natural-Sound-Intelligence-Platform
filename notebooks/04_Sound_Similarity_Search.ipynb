```python
import pandas as pd
import numpy as np
from pathlib import Path

from sklearn.preprocessing import StandardScaler
from sklearn.metrics.pairwise import euclidean_distances
```

```python
PROJECT_ROOT = Path.cwd().parent

FEATURES_PATH = (
    PROJECT_ROOT /
    "outputs" /
    "natural_sound_features.csv"
)

features_df = pd.read_csv(FEATURES_PATH)

print(features_df.shape)

features_df.head()
```

```python
feature_columns = [

    "mean",
    "std",
    "min",
    "max",

    "skewness",
    "kurtosis",

    "rms",

    "entropy",

    "zcr",

    "spectral_centroid",
    "spectral_bandwidth",
    "spectral_rolloff"

]

X = features_df[feature_columns]

print(X.shape)
```

```python
scaler = StandardScaler()

X_scaled = scaler.fit_transform(X)

print(X_scaled.shape)
```

```python
distance_matrix = euclidean_distances(X_scaled)

print(distance_matrix.shape)
```

```python
distance_matrix[:5, :5]
```

```python
def find_similar_sounds(index, top_n=5):

    distances = distance_matrix[index]

    nearest_indices = np.argsort(distances)

    nearest_indices = nearest_indices[1:top_n+1]

    results = features_df.iloc[
        nearest_indices
    ][["filename", "category"]]

    return results
```

```python
find_similar_sounds(0)
```
